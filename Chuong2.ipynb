```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
image_paths = [
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/bisons.jpg",
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/aero.jpg",
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/horse.jpg"
]
plt.figure(figsize=(15, 10))
for i, path in enumerate(image_paths):
    img = cv2.imread(path, 0)
    dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
    dft_shift = np.fft.fftshift(dft)
    magnitude_spectrum = cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1])


    plt.subplot(3, 2, 2 * i + 1)  # Lưới 3x2: cột 1
    plt.imshow(img, cmap='gray')
    plt.title(f'Original Image {i + 1}')
    plt.axis('off')

    plt.subplot(3, 2, 2 * i + 2)
    plt.imshow(np.log(magnitude_spectrum + 1), cmap='gray')
    plt.title(f'Magnitude Spectrum {i + 1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
