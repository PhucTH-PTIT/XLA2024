```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image_paths = [
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/me.jpg",
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/me4.jpg",
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/me5.jpg"
]
plt.figure(figsize=(15, 10))

for idx, path in enumerate(image_paths):
    img = cv2.imread(path, 0)
    if img is None:
        print(f"Error: Cannot read image from path: {path}")
        continue
    sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
    edges = cv2.magnitude(sobel_x, sobel_y)

    plt.subplot(len(image_paths), 3, 3 * idx + 1)
    plt.imshow(img, cmap='gray')
    plt.title(f'Original Image {idx + 1}')
    plt.axis('off')

    plt.subplot(len(image_paths), 3, 3 * idx + 2)
    plt.imshow(edges, cmap='gray')
    plt.title(f'Edges {idx + 1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
