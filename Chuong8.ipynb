```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Đường dẫn tới các ảnh
image_paths = [
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/horse.jpg",
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/messi.jpg",
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/ronaldo.jpg"
]
plt.figure(figsize=(10, 10))

for idx, path in enumerate(image_paths):
    img = cv2.imread(path, 0)
    if img is None:
        print(f"Error: Cannot read image from path: {path}")
        continue
    ret, thresh_otsu = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    plt.subplot(3, 2, 2 * idx + 1)  # 3 hàng, 2 cột, ảnh gốc
    plt.imshow(img, cmap='gray')
    plt.title(f'Original Image {idx + 1}')
    plt.axis('off')

    plt.subplot(3, 2, 2 * idx + 2)  # 3 hàng, 2 cột, ảnh sau Otsu Thresholding
    plt.imshow(thresh_otsu, cmap='gray')
    plt.title(f'Otsu Thresholding {idx + 1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
