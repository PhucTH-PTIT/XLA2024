```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image_paths = [
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/lena.jpg",
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/man.jpg",
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/messi.jpg"
]

plt.figure(figsize=(15, 5))

for idx, path in enumerate(image_paths):
    img = cv2.imread(path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    if img is None:
        print(f"Error: Cannot read image from path: {path}")
        continue
    dst = cv2.cornerHarris(np.float32(gray), 2, 3, 0.04)
    dst = cv2.dilate(dst, None)

    img[dst > 0.01 * dst.max()] = [0, 0, 255]

    # Hin th kt qu (1 hng, 3 ct)
    plt.subplot(1, 3, idx + 1)  # 1 hng, 3 ct
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Chuyn sang RGB  hin th
    plt.title(f'Harris Corner {idx + 1}')
    plt.axis('off')

plt.tight_layout()
plt.show()

```
