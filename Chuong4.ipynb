```python
import cv2
import matplotlib.pyplot as plt

# ng dn ti cc nh
image_paths = [
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/mandrill_spnoise_0.2.jpg",
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/mandrill_spnoise_0.3.jpg",
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/beans_g.png"
]

plt.figure(figsize=(15, 10))

for idx, path in enumerate(image_paths):
    img = cv2.imread(path, 0)
    if img is None:
        print(f"Error: Cannot read image from path: {path}")
        continue
    equ = cv2.equalizeHist(img)

    plt.subplot(len(image_paths), 2, 2 * idx + 1)
    plt.imshow(img, cmap='gray')
    plt.title(f'Original Image {idx + 1}')
    plt.axis('off')

    plt.subplot(len(image_paths), 2, 2 * idx + 2)
    plt.imshow(equ, cmap='gray')
    plt.title(f'Equalized Image {idx + 1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
