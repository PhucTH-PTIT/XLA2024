```python
from PIL import Image
import matplotlib.pyplot as plt
image_paths = [
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/aero.jpg",
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/cameraman.jpg",
    "C:/Users/ngtrd/Desktop/Sandipan_Dey_2018_Sample_Images/images/bisons.jpg"
]
images = [Image.open(img_path) for img_path in image_paths]
plt.figure(figsize=(15, 5))
for i, img in enumerate(images):
    plt.subplot(1, 3, i + 1) 
    plt.imshow(img)
    plt.axis('off')  
    plt.title(f"Image {i + 1}")

plt.show()
```
